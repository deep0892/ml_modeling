```python
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/drive/MyDrive/Plaksha/Term 7/Data Visualization/Butin_antibiotic_data.xlsx')
df.head()
```

```python
df_positive = df[df['Gram Staining'] == 'positive']
df_negative = df[df['Gram Staining'] == 'negative']
```

**Initial Attempt for Visualization**

```python
x = [-4. , -3. , -2.3 , -2. , -1.52 ,  -1. , -0.3979 ,   0. , 0.3 , 0.6989 ,  1. , 2. , 3. ,  4]

fig=plt.figure(figsize=(15, 8))
fig.suptitle("Burtin's Antibiotics", fontsize=16)

plt.subplot(2, 1, 1)
plt.scatter(
    np.log10(df_positive['Penicilin']), 
    df_positive['Bacteria'],
    alpha=0.5,
    label='Penicilin - Gram positive',
    color='blue'
)

plt.scatter(
    np.log10(df_positive['Streptomycin']), 
    df_positive['Bacteria'], 
    alpha=0.5,
    label='Streptomycin - Gram positive',
    color='orange'
)

plt.scatter(
    np.log10(df_positive['Neomycin']),
    df_positive['Bacteria'], 
    alpha=0.5,
    label='Neomycin - Gram positive',
    color='red'
)

plt.ylabel("Bacteria", size=14)
plt.grid(which='both',axis='both', linewidth=0.5)

plt.xticks(
    ticks = x,
    labels = np.round(np.power(10, x),4)
) 

plt.legend()
#--------------------------------#
# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.scatter(
    np.log10(df_negative['Penicilin']), 
    df_negative['Bacteria'],
    alpha=0.5,
    label='Penicilin- Gram negative',
    color='blue'
)

plt.scatter(
    np.log10(df_negative['Streptomycin']), 
    df_negative['Bacteria'], 
    alpha=0.5,
    label='Streptomycin - Gram negative',
    color='orange'
)

plt.scatter(
    np.log10(df_negative['Neomycin']),
    df_negative['Bacteria'], 
    alpha=0.5,
    label='Neomycin - Gram negative',
    color='red'
)

plt.xlabel("MIC", size=14)
plt.ylabel("Bacteria", size=14)

plt.grid(which='both',axis='both', linewidth=0.5)

plt.xticks(
    ticks = x,
    labels = np.round(np.power(10, x),4)
)

plt.legend()

fig.tight_layout()
fig.subplots_adjust(top=0.9)
plt.show()
```

**Final Visualization**

```python
x = [-4. , -3. , -2.3 , -2. , -1.52 ,  -1. , -0.3979 ,   0. , 0.3 , 0.6989 ,  1. , 2. , 3. ,  4]

fig=plt.figure(figsize=(15, 8))
fig.suptitle("Burtin's Antibiotics", fontsize=16)
plt.subplot(1, 2, 1)

plt.plot(
    df_positive['Bacteria'],
    np.log10(df_positive['Penicilin']), 
    ':o',
    label='Penicilin-positive',
    color='blue'
)
plt.plot(
    df_positive['Bacteria'], 
    np.log10(df_positive['Streptomycin']), 
    ':o',
    label='Streptomycin-positive',
    color='orange'
)

plt.plot(
    df_positive['Bacteria'], 
    np.log10(df_positive['Neomycin']),
    ':o',
    label='Neomycin-positive',
    color='red'
)

plt.ylabel("MIC", size=14)
plt.xlabel("Bacteria", size=14)
plt.grid(which='both',axis='both', linewidth=0.5)

plt.yticks(
    ticks = x,
    labels = np.round(np.power(10, x),4)
) 
plt.title("Gram Staining - Positive")
plt.xticks(rotation=90)
plt.legend()

#--------------------------------#
plt.subplot(1, 2, 2)
plt.plot(
    df_negative['Bacteria'],
    np.log10(df_negative['Penicilin']), 
    ':o',
    label='Penicilin-negative',
    color='blue'
)

plt.plot( 
    df_negative['Bacteria'], 
    np.log10(df_negative['Streptomycin']),
    ':o',
    label='Streptomycin-negative',
    color='orange'
)

plt.plot(
    df_negative['Bacteria'], 
    np.log10(df_negative['Neomycin']),
    ':o',
    label='Neomycin-negative',
    color='red'
)

plt.title("Gram Staining - Negative")
plt.xlabel("Bacteria", size=14)

plt.grid(which='both',axis='both', linewidth=0.5)
plt.xticks(rotation=90)
plt.legend()

frame1 = plt.gca()
frame1.axes.yaxis.set_ticklabels([])
plt.yticks(
    ticks = x,
)

fig.tight_layout()
fig.subplots_adjust(top=0.88)
plt.show()
```

Comments:




*   First I tried to visualize the plot using scatter plot to show bubble for each Antibiotics and then plot the same for the gram positive and negative. The plot didnt seem intuitive enough to make any inferences.
*   This gave me the idea to plot a line plot with dot for points which would give a better understanding of the plot for different antibiotics.
*   Further removing the y axis labels and having the same axis lines helps in visualizing the 2 plots together and make a comparison.
*   Making 2 subplots for positive and negative gram staining made sense as ploting the same together would cluter the graph.
